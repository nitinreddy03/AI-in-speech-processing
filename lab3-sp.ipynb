```python
import librosa
import soundfile as sf
from IPython.display import Audio, display
 
# Load the recorded speech signal
recorded_signal, sr = librosa.load(r"C:\Users\91939\Downloads\lab3.mp3", sr=None)
 
# A1. Remove silence from beginning and end of the signal
trimmed_signal, index = librosa.effects.trim(recorded_signal)
 
# Play the trimmed signal
display(Audio(trimmed_signal, rate=sr))
 
# A2. Split the recorded speech with detected silences
split_signals = librosa.effects.split(recorded_signal, top_db=30)  # Adjust top_db parameter as needed
 
# Play each split signal
for i, split in enumerate(split_signals):
    split_signal = recorded_signal[split[0]:split[1]]
    display(Audio(split_signal, rate=sr))

```
