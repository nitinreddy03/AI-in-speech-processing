```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import IPython.display as ipd

new_audio_file = 'Recording1.wav'

y_new, sr_new = librosa.load(new_audio_file)

derivative_1_new = np.diff(y_new)
derivative_1_new /= np.max(np.abs(derivative_1_new))

print("Playing First Derivative Signal (New):")
plt.figure(figsize=(10, 5))
librosa.display.waveshow(derivative_1_new, sr=sr_new, color='green')
plt.title('FIRST DERIVATIVE (New)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()

ipd.Audio(derivative_1_new, rate=sr_new)
```

```python
new_audio_signal = y_new  # Use the new audio signal you loaded
new_audio_rate = sr_new    # Use the sample rate of the new audio signal

# Display the original audio signal
plt.figure(figsize=(10, 5))
librosa.display.waveshow(new_audio_signal, sr=new_audio_rate, color='blue')
plt.title('ORIGINAL')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()

# Play the original audio signal
ipd.Audio(new_audio_signal, rate=new_audio_rate)
```

```python


import librosa
import numpy as np
import matplotlib.pyplot as plt

person_1 = r"C:\Users\vikra\Downloads\nitin.mp3"
person_2 = r"C:\Users\vikra\Downloads\surya.mp3"


# Load audio signals
y1, sr1 = librosa.load(person_1)
y2, sr2 = librosa.load(person_2)

# Duration of audio files
duration1 = librosa.get_duration(y=y1, sr=sr1)
duration2 = librosa.get_duration(y=y2, sr=sr2)

print("The length Duration of first audio file:", duration1, "seconds")
print("The length Duration of Second audio file:", duration2, "seconds")

# Remove silence function
def remove_silence(y, sr, threshold=0.01):
    yt = librosa.effects.trim(y, top_db=threshold)
    return yt[0]

# Trim silence from audio signals
person_1_trimmed = remove_silence(y1, sr1)
person_2_trimmed = remove_silence(y2, sr2)

# Calculate the time axes for the trimmed audio files
time1 = np.linspace(0, len(person_1_trimmed) / sr1, len(person_1_trimmed))
time2 = np.linspace(0, len(person_2_trimmed) / sr2, len(person_2_trimmed))

# Plot the trimmed audio files
plt.figure(figsize=(16, 6))
plt.plot(time1, person_1_trimmed, label='person_1')
plt.plot(time2, person_2_trimmed, label='person_2')
plt.legend()
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Overlapped Audio Files')
plt.show()
```
